Saffron UK import data: 2019 - 2023 provided by: https://wits.worldbank.org/trade/comtrade/en/country/GBR/year/2022/tradeflow/Imports/partner/ALL/product/091020

```python
import pandas as pd

uk_import_path_2019 = "./uk-imports-data/uk-import-2019.xlsx"
uk_import_path_2020 = "./uk-imports-data/uk-import-2020.xlsx"
uk_import_path_2021 = "./uk-imports-data/uk-import-2021.xlsx"
uk_import_path_2022 = "./uk-imports-data/uk-import-2022.xlsx"
uk_import_path_2023 = "./uk-imports-data/uk-import-2023.xlsx"

imports_2019 = pd.ExcelFile(uk_import_path_2019)
imports_2020 = pd.ExcelFile(uk_import_path_2020)
imports_2021 = pd.ExcelFile(uk_import_path_2021)
imports_2022 = pd.ExcelFile(uk_import_path_2022)
imports_2023 = pd.ExcelFile(uk_import_path_2023)

dfs_19 = {sheet_name: imports_2019.parse(sheet_name) 
          for sheet_name in imports_2019.sheet_names}

dfs_20 = {sheet_name: imports_2020.parse(sheet_name) 
          for sheet_name in imports_2020.sheet_names}

dfs_21 = {sheet_name: imports_2021.parse(sheet_name) 
          for sheet_name in imports_2021.sheet_names}

dfs_22 = {sheet_name: imports_2022.parse(sheet_name) 
          for sheet_name in imports_2022.sheet_names}

dfs_23 = {sheet_name: imports_2023.parse(sheet_name) 
          for sheet_name in imports_2023.sheet_names}


data_19 = dfs_19["By-HS6Product"]
data_20 = dfs_20["By-HS6Product"]
data_21 = dfs_21["By-HS6Product"]
data_22 = dfs_22["By-HS6Product"]
data_23 = dfs_23["By-HS6Product"]
```

```python
data_19 = pd.DataFrame(data_19,
                  columns=['Reporter', 'TradeFlow', 'ProductCode', 'Product Description', 'Year', 'Partner', 'Trade Value 1000USD', 'Quantity', 'Quantity Unit'])

data_19 = data_19.drop(columns=["Reporter", "TradeFlow", "ProductCode", "Product Description", "Quantity Unit"])

data_20 = pd.DataFrame(data_20,
                  columns=['Reporter', 'TradeFlow', 'ProductCode', 'Product Description', 'Year', 'Partner', 'Trade Value 1000USD', 'Quantity', 'Quantity Unit'])

data_20 = data_20.drop(columns=["Reporter", "TradeFlow", "ProductCode", "Product Description", "Quantity Unit"])

data_21 = pd.DataFrame(data_21,
                  columns=['Reporter', 'TradeFlow', 'ProductCode', 'Product Description', 'Year', 'Partner', 'Trade Value 1000USD', 'Quantity', 'Quantity Unit'])

data_21 = data_21.drop(columns=["Reporter", "TradeFlow", "ProductCode", "Product Description", "Quantity Unit"])

data_22 = pd.DataFrame(data_22,
                  columns=['Reporter', 'TradeFlow', 'ProductCode', 'Product Description', 'Year', 'Partner', 'Trade Value 1000USD', 'Quantity', 'Quantity Unit'])

data_22 = data_22.drop(columns=["Reporter", "TradeFlow", "ProductCode", "Product Description", "Quantity Unit"])

data_23 = pd.DataFrame(data_23,
                  columns=['Reporter', 'TradeFlow', 'ProductCode', 'Product Description', 'Year', 'Partner', 'Trade Value 1000USD', 'Quantity', 'Quantity Unit'])

data_23 = data_23.drop(columns=["Reporter", "TradeFlow", "ProductCode", "Product Description", "Quantity Unit"])

data_19 = data_19.rename(columns={"Partner": "From", "Trade Value 1000USD": "value"})
data_20 = data_20.rename(columns={"Partner": "From", "Trade Value 1000USD": "value"})
data_21 = data_21.rename(columns={"Partner": "From", "Trade Value 1000USD": "value"})
data_22 = data_22.rename(columns={"Partner": "From", "Trade Value 1000USD": "value"})
data_23 = data_23.rename(columns={"Partner": "From", "Trade Value 1000USD": "value"})
```

Dataframe by baci: https://www.cepii.fr/CEPII/en/bdd_modele/bdd_modele_item.asp?id=37 HS22 (2022), functions below only obtain saffron for uk

```python
def map_country_codes(country_codes_path):
    country_codes_df = pd.read_csv(country_codes_path)
    country_codes_map = {}

    for _, r in country_codes_df.iterrows():
        country_codes_map[r["country_code"]] = r["country_name"]
    
    return country_codes_map

def get_saffron_imports(data_path, country_codes_map):
    df_baci = pd.read_csv(data_path)

    df_baci.rename(columns={"t": "year", "i": "exporter", "j": "importer", "k": "product", "v": "value", "q": "quantity"}, inplace=True)

    saffron_data = df_baci[df_baci['product'] == 91020] #saffron code is 91020
    saffron_data_uk = saffron_data[saffron_data['importer'] == 826] #uk code 826

    for i, r in saffron_data_uk.iterrows():
        saffron_data_uk.at[i, "exporter"] = country_codes_map[r["exporter"]]
    
    return saffron_data_uk
```

```python
path_baci_2022 = "./baci_data/BACI_HS22_Y2022_V202401b.csv"
country_codes_path = "./baci_data/country_codes_V202401b.csv"

country_codes_map = map_country_codes(country_codes_path)
data_saffron_uk_2022 = get_saffron_imports(path_baci_2022, country_codes_map)

data_saffron_uk_2022
```
